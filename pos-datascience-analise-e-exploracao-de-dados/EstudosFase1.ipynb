```python
#Para solucionar o problema dos ticks e facilitar análises futuras, vou converter as datas para o formato datetime (são os headers das colunas)



#Criando um dicionário para mapear os meses
meses_mapeados = {
    'Jan': '01', 'Fev': '02', 'Mar': '03', 'Abr': '04', 'Mai': '05', 'Jun': '06',
    'Jul': '07', 'Ago': '08', 'Set': '09', 'Out': '10', 'Nov': '11', 'Dez': '12'
}

#Função para converter o header, dado o padrão "AAAA/Mon"
def conversor(header):
    try:
        #Verifica se o header possui a barra separadora

        if '/' in header:
            ano, mes_str = header.split('/')

            #Mapeia o mês abreviado para seu número
            mes = meses_mapeados.get(mes_str)
            if mes is None:
                return header  # se o mês não for reconhecido, retorna o header original
            
            #Converte para objeto datetime assumindo o dia 1
            return datetime.strptime(f'{ano}-{mes}-01', '%Y-%m-%d')
        
        else:
            return header
    except Exception as e:
        #Em caso de erro, retorna o header original
        return header

#Transformando as colunas do DataFrame
dados_ordenado.columns = [conversor(coluna) for coluna in dados_ordenado.columns]


```

```python
#Inserindo agora informações demográficas do IBGE para fazer uma análise mais completa
#dados demográficos do IBGE

populacao = pd.read_csv('C:\\Programacao\\dados\\tabelapopulacaobr.csv', sep=";", thousands=".", decimal=",", skiprows = 4, skipfooter=21, engine ='python')
populacao
```

```python
#Vou limpar os índices de número e espaço antes de cada estado no dados_ordenado para conseguir fazer o merge depois
#O padrão que quero remover é uma sequência de dígitos seguidos de um ou mais espaços no início da string. A expressão regular para isso seria r'^\d+\s+'
#significados: 
# ^ indica o início da string.
# \d+ corresponde a um ou mais dígitos.
# \s+ corresponde a um ou mais espaços.
#regex é regular expression, nesse caso é a expressão que se repete que eu quero tirar de todas as strings

dados_ordenado.index = dados_ordenado.index.str.replace(r'^\d+\s+', '', regex=True)


#Agora dá para dar merge!
```

```python
#Vou criar uma função que recebe de entrada a data específica e retorna um novo df que contém as colunas: UF, população, valor gasto e gasto per capita!

def gasto_per_capita(data):
    #pega a data desejada como timestamp
    data_ts = pd.Timestamp(data)

    #copia a coluna do gasto escolhido para uma nova variável
    df_gasto = dados_ordenado[[data_ts]].copy()

    #como o índice é a UF, vou trasnformar ele em uma coluna, pra poder dar merge com o df novo
    df_gasto.reset_index(inplace=True)

    #renomeando a coluna para um nome mais intuitivo na hora de mostrar a tabela
    df_gasto.rename(columns={data_ts: 'Valor Gasto'}, inplace=True)

    #dando merge dos dois dfs
    df_final = pd.merge(df_gasto, populacao, on='Unidade da Federação', how='left')

    #criando uma nova coluna que é o gasto per capita (dividindo os gastos de cada UF pela população)
    df_final['Gasto per Capita'] = df_final['Valor Gasto'] / df_final['Total']

    #tornando as UFs como índice novamente
    df_final.set_index('Unidade da Federação', inplace=True)

    #removendo a coluna # que aparece do nada.
    df_final.drop(columns='#', inplace=True, errors='ignore')

    return df_final
    
    

#O censo usado foi o de 2022, que foi feito em agosto desse ano. Então, vou usar os dados dessa data para fazer a análise mais fidedigna possível e os gráficos.

data_especifica = '2022-08-01'
df_gasto = gasto_per_capita(data_especifica)
df_gasto 
```

```python
#Dando sort no df com base na coluna 'Gasto per Capita' para ver quais estados gastaram mais por pessoa 
df_gasto.sort_values(by='Gasto per Capita', ascending=False, inplace=True)

df_gasto
```

```python
#novamente, gerando um gráfico:

#criando a figura e os eixos
fig, ax = plt.subplots(figsize=(12, 6))

#destacando Minas Gerais com uma cor diferente para fins comparativos
cores = ['darkred' if estado == 'Minas Gerais' else '#3498DB' for estado in df_gasto.index]

#plotando

plt.bar(df_gasto.index, df_gasto["Gasto per Capita"], color=cores)


#adicionando uma linha horizontal na ponta da barra de Minas Gerais para que fique mais fácil de comparar com os outros estados

#preciso especificar ["Minas Gerais", "Gasto per Capita"] aqui. 
#O problema é que, ao usar df_gasto.loc['Minas Gerais'] sem especificar uma coluna, está retornando 
#toda a linha referente a "Minas Gerais", ou seja, uma Série com vários valores. A função ax.axhline() 
#espera um valor numérico único para o parâmetro y, e não uma Série.
if 'Minas Gerais' in df_gasto.index:
    ax.axhline(y = df_gasto.loc["Minas Gerais", "Gasto per Capita"], color='darkred', linestyle='dashdot', linewidth=0.5)


#configurando o label do eixo x
for label in ax.get_xticklabels():
    label.set_rotation(45)
    label.set_ha('right')

#configurando o layout
#ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x/1e6:,.0f}'))
ax.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.45)
#ax.set_ylim(0, 450000000)
ax.set_xlabel('Unidade da Federação')
ax.set_ylabel('Valor per Capita (Em Reais)')
ax.set_title(f'Gasto per Capita por Estado (SUS) em {data_especifica}')

plt.show()



#para fazer o valor per capita por estado por ano, por exemplo, seria fazer um sum dos meses de um ano pra fazer uma coluna "Total" para aquele ano ou 
#período de tempo específico e depois fazer o mesmo processo que tá aqui.
```
